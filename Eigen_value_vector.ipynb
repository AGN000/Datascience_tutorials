```python
import numpy as np
import matplotlib.pyplot as plt

# Define a transformation matrix
A = np.array([[3, 1],
              [0, 2]])

# Compute eigenvalues and eigenvectors
eigvals, eigvecs = np.linalg.eig(A)

# Pick one eigenvector
v = eigvecs[:, 0]
lambda_val = eigvals[0]
Av = A @ v   # should be lambda * v

# Pick a random non-eigenvector
u = np.array([1, 2])   # arbitrary vector not aligned with eigenvector
Au = A @ u

# Plot setup
plt.figure(figsize=(8,8))
ax = plt.gca()

# Eigenvector before and after transformation
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='black', label='Eigenvector v')
plt.quiver(0, 0, Av[0], Av[1], angles='xy', scale_units='xy', scale=1, color='red', label=r'Transformed $\lambda v$')

# Non-eigenvector before and after transformation
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Non-eigenvector u')
plt.quiver(0, 0, Au[0], Au[1], angles='xy', scale_units='xy', scale=1, color='green', label='Transformed Au')

# Axes setup
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal', adjustable='box')

plt.legend()
plt.title("Eigenvector vs Non-Eigenvector under Transformation")
plt.show()
```
