```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Grid
# -----------------------------
grid_points = 200
x = np.linspace(-2, 2, grid_points)
y = np.linspace(-2, 2, grid_points)
X, Y = np.meshgrid(x, y)

# -----------------------------
# Functions
# -----------------------------
Z_skewed = X**2 + 100*Y**2   # skewed
Z_symmetric = X**2 + Y**2    # symmetric

# -----------------------------
# Plot contours
# -----------------------------
fig, axs = plt.subplots(1, 2, figsize=(12,5))

cs0 = axs[0].contourf(X, Y, Z_skewed, levels=50, cmap='viridis')
axs[0].set_title("Skewed Function: $x^2 + 100 y^2$")
fig.colorbar(cs0, ax=axs[0])

cs1 = axs[1].contourf(X, Y, Z_symmetric, levels=50, cmap='viridis')
axs[1].set_title("Symmetric Function: $x^2 + y^2$")
fig.colorbar(cs1, ax=axs[1])

for ax in axs:
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.tight_layout()
plt.show()
```
