# Python Tutorial Notebook
This tutorial covers loops, functions, conditional statements, PIP, virtual environments, input handling, and file handling.

## 1. Loops in Python

````
# For loop example
for i in range(5):
    print("Iteration", i)

# While loop example
count = 0
while count < 3:
    print("Count:", count)
    count += 1
````

## 2. Conditional Statements

````
x = 10
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")
````

## 3. Functions in Python

````
def greet(name):
    return f"Hello, {name}!"

print(greet("Arun"))

# Function with default arguments
def add(a, b=5):
    return a + b

print(add(10))  # Uses default b
````

## 4. PIP Usage

Install a package:
```
pip install numpy
```
List installed packages:
```
pip list
```

## 5. Virtual Environment

Create a virtual environment:
```
python -m venv myenv
```
Activate it (Windows):
```
myenv\Scripts\activate
```
Linux/macOS:
```
source myenv/bin/activate
```

## 6. Input Handling

````
name = input("Enter your name: ")
print("You entered:", name)
````

## 7. File Handling

````
# Writing to a file
with open("example.txt", "w") as f:
    f.write("Hello, file!")

# Reading from a file
with open("example.txt", "r") as f:
    data = f.read()
    print(data)
````
