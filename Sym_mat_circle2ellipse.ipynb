```python
import numpy as np
import matplotlib.pyplot as plt

# Define a symmetric matrix (so we get real eigenvalues and orthogonal eigenvectors)
A = np.array([[3, 5],
              [5, 2]])

# Eigen decomposition
eigvals, eigvecs = np.linalg.eig(A)

# Create a unit circle
theta = np.linspace(0, 2*np.pi, 200)
circle = np.array([np.cos(theta), np.sin(theta)])  # shape (2, N)

# Apply transformation
ellipse = A @ circle

# Plot
plt.figure(figsize=(7,7))

# Original circle
plt.plot(circle[0], circle[1], 'b--', label='Unit Circle')

# Transformed ellipse
plt.plot(ellipse[0], ellipse[1], 'r', label='Transformed Ellipse')

# Draw eigenvectors scaled by eigenvalues (major/minor axes)
for i in range(len(eigvals)):
    vec = eigvecs[:, i] * eigvals[i]   # scale eigenvector by eigenvalue
    plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, 
               width=0.01, color='k', label=f"Eigenvector {i+1}")

# Axes setup
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Circle  Ellipse under Linear Transformation\nEigenvectors = directions, Eigenvalues = axis lengths")
plt.show()
```
